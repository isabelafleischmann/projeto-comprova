Programa para extrair textos de checagens etiquetadas como "Enganosas" no Projeto Comprova

```python
import requests #para enviar solicitações https
from bs4 import BeautifulSoup as bs #para analisar conteúdo html 
```

```python
url = "https://projetocomprova.com.br/"  #mandando um request pro site alvo para receber o conteúdo como resposta
resposta = requests.get(url) #pra isso usamos a função "get"
html = resposta.text
```

```python
soup = bs(html, 'html.parser') #passando o conteúdo recebido para o formato HTML usando o BeautifulSoup
```

```python
# os artigos estão em tags article e, com o BeautifulSoup, conseguimos uma lista deles
artigos = soup.find_all('article')
```

```python
artigos[0]
```

```python
# pegando os enganosos
artigos_enganosos = []
for artigo in artigos:
  if 'Enganoso' in str(artigo):
    artigos_enganosos.append(artigo)
```

```python
len(artigos_enganosos)
```

```python
#Conseguindo o texto dos artigos enganosos
```

```python
artigos_enganosos[0].text.splitlines()
```

```python
conteudos = []

for artigo in artigos_enganosos:
  # dentro de cada artigo, o texto está em tags section
  sections = artigo.find_all('section')
  intro = sections[0].text
  texto = sections[1].text
  # pegando também título e link
  titulo = artigo.find('h1', {'class': 'answer__title'}).text
  link = artigo.find('a', {'class': 'answer__title__link'}).attrs['href']
  dicio = {
      'título': titulo,
      'link': link,
      'introdução': intro,
      'artigo': texto
  }
  conteudos.append(dicio)
```

```python
conteudos[0]
```

```python
import pandas as pd
```

```python
#df no pandas pra salvar os dados como csv
```

```python
df = pd.DataFrame(conteudos)
df.to_csv('dados_projeto_comprova_enganosos.csv', index= False)
```
